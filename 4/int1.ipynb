```python
import pandas as pd
import numpy as np
```

```python
source_data = pd.read_csv('../data/source_data.csv', index_col=0)
source_data
```

```python
NRR = source_data['Number of  reported results']
# 做差分并绘制图像
NRR_diff = NRR.diff()
NRR_diff = NRR_diff / NRR
NRR_diff = abs(NRR_diff)
NRR_diff.plot()
```

```python
# NRR_diff 标准化使在 0-1 之间
NRR_diff = (NRR_diff - NRR_diff.min()) / (NRR_diff.max() - NRR_diff.min())
NRR_diff.to_list()
```
